# Retail Warehouse — Analytics Exploration

Run `python -m src.pipeline` (or `make pipeline`) first to populate `data/warehouse/retail.db`.

```python
import sys
sys.path.append('..')

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect('../data/warehouse/retail.db')
```

```python
daily = pd.read_sql('SELECT * FROM mart_daily_sales_by_store', conn)
daily.groupby('date_key')['net_revenue'].sum().plot(figsize=(10,4), title='Total Daily Net Revenue')
plt.show()
```

```python
monthly = pd.read_sql('SELECT * FROM mart_monthly_sales_by_category', conn)
monthly.groupby('category')['gross_profit'].sum().sort_values(ascending=False).plot(kind='bar', title='Total Gross Profit by Category')
plt.show()
```

```python
top_customers = pd.read_sql('SELECT * FROM mart_customer_summary ORDER BY lifetime_revenue DESC LIMIT 10', conn)
top_customers
```

```python
dq = pd.read_sql('SELECT dq_reasons FROM stg_quarantined_transactions', conn)
dq['dq_reasons'].str.split('; ').explode().value_counts().plot(kind='barh', title='Quarantined Rows by Reason')
plt.show()
```
